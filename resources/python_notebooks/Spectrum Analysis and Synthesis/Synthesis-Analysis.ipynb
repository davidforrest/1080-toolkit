## Review of The Fourier Series <a class="anchor" id="review_FS"></a>
[Joseph Fourier](http://lpsa.swarthmore.edu/Fourier/Series/FourierBio.html) was a French mathmetician whose interest in heat transfer and vibrations brought us what is called Fourier Analysis. We are often concerned with Fourier analysis for Frequency spectrum analysis, but Fourier came up with the Fourier Series expansion of a periodic function in his search for solutions to the heat equation, which is a partial differential equation that describes the spatial distribution of heat in a medium. As we saw with solving the wave equation, if it can be solved in the case of a single sinusoid, then using the superposition principle for this linear equation, the solution for an arbitrary periodic function is immediately available by expressing the original function as a sum of sinuoids (a Fourier Series) and then plugging in the solution for each sinusoidal component. 

The Fourier Series is a recipe for constructing a periodic signal, with precise amounts of each ingredient ($\sin$, $\cos$ or complex exponential functions). It is a series expansion of a periodic function $x(t)$ and it requires that the function be absolutely summable $\int_{-\infty}^{\infty} |x(t)| \hspace{0.1cm} dt < \infty$. Here is a [tool](http://www.falstad.com/fourier/Fourier.html) that helps visualize this concept of *ingredients* with animations.

**Synthesis**

The Fourier Series synthesis equations describe how to synthesize or build a periodic function of basic building blocks and it has two forms:

**The trigonometric series form**
\begin{equation*}
x(t) = a_0 + \sum_{n=1}^{\infty} a_n cos(n\omega_o t) + b_n sin(n\omega_o t) \\
\end{equation*}

**The exponential series form**
\begin{equation*}
x(t) = \sum_{n=\infty}^{\infty} c_n e^{jn\omega_ot}
\end{equation*}

The process of ** Analysis ** has to do with calulating the coefficients $a_n$, $b_n$ for a given signal. You are given the cake, now apply the tools to tell you what the recipe or the ingredients are. For an interesting discussion on that, refer to the [Appendix](#Appendix) below.

## Fundamental, partial and harmonic review <a class="anchor" id="fun_part_har"></a>

The lowest frequency component is called the *Fundamental frequency*. Sometimes the fundamental frequency is also the frequency with the highest amplitude. Then it's a dominant frequency. We call multiples of the fundamental harmonics because they are harmonious with the fundamental. The spectrum of some instruments could also contain other peaks that characterise the sound or the instrument, but that are not necessarily multiples of the fundamental. We refer to them as partials in that case. 

From Loy vol.1 "The individual sinusoids that collectively make up an instrumental tone are called its partials because each carries a partial characterization of the whole sound. Partials are also known as
components, and I will use these terms interchangeably. The principal properties of the partials are their frequencies and amplitudes. The way the partials manifest in frequency, amplitude, and time is what our ears use to determine what kind of instrument made a particular sound.

The lowest pitched partial in a tone is called the fundamental. It is generally what our ears pick out as the pitch of the tone. Since, by definition, the remaining partials in the tone are pitched higher, they are called overtones. Our ears use the pattern of overtone frequencies as an important cue to recognize timbres. The overtone frequencies of wind and string instruments are positive integer multiples of the fundamental, where the positive integers are 1, 2, 3, and so on. For instance, if a flute or violin has fundamental frequency f, then the frequencies of its overtones will be positive integer multiples of f. The partials of such instruments are called harmonics. Note that because the positive numbers start at 1, and because 1 × f = f, therefore the first harmonic is the same as the fundamental. Instruments with harmonic partials are usually chosen to carry the melody and harmony of
music because frequencies of the harmonics tend to agree in frequency with the pitches of the diatonic scale. Instruments with inharmonic partials such as drums and bells are usually not used to carry melody and harmony because for the most part the frequencies of their partials do not agree with the diatonic scale.
The amplitudes and frequencies of the partials of musical instruments tend to vary in a characteristic way over the duration of a tone, depending upon the instrument and performance style of the performer. Though the variation may be slight, the precise amplitude and frequency ballistics of the partials help our ears to fuse a single percept of an instrument out of its individual partials, and help identify the type of instrument."

From Muller (Ch1): "Real-world sounds are far from being a simple pure tone with a well-defined frequency. Playing a single note on an instrument may result in a complex sound that contains a mixture of different frequencies changing over time. Intuitively, such a
musical tone can be described as a superposition of pure tones or sinusoids, each with its own frequency of vibration, amplitude, and phase. A partial is any of the sinusoids by which a musical tone is described. The frequency of the lowest partial present is called the fundamental frequency of the sound. The pitch of a musical
tone is usually determined by the fundamental frequency, which is the one created by vibration over the full length of a string or air column of an instrument. A harmonic (or a harmonic partial) is a partial that is an integer multiple of the fundamental frequency. Partials, as well as harmonics, are counted upwards along the frequency axis. This convention implies that the fundamental frequency is the
first partial, as well as the first harmonic of a musical tone. The term inharmonicity is used to denote a measure of the deviation of a partial from the closest ideal harmonic, typically measured in cents for each partial. Finally, another term that is often used in music theory is the overtone, which is any partial except the lowest.
This can lead to numbering confusion when comparing overtones with partials, since the first overtone is the second partial.
Most pitched instruments are designed to have partials that are close to being harmonics, with very low inharmonicity. Thus, for simplicity, one often speaks of the partials in those instruments’ sounds as harmonics, even if they have some inharmonicity.
Other pitched instruments, especially certain percussion instruments, such as the marimba, vibraphone, bells, and kettledrums (timpani), contain nonharmonic partials, yet give the ear a good sense of pitch. Nonpitched, or indefinite-pitched, instruments, such as cymbals, gongs, or tam-tams, make sounds rich in inharmonic partials."

## Review of the Fourier Transform

As briefly discussed in class, when the period $T$ of the above function $x(t)$ is allowed to go to $\infty$, in otherwords we move to signals that are not necessarily periodic, we get the Fourier Transform Pair, and the synthesis and analysis equations as follows:

#### Synthesis
\begin{equation*}
x(t) = {1 \over 2 \pi} \int_{-\infty}^{\infty} X(\omega) \hspace{0.1cm} e^{-j\omega t} d\omega 
\end{equation*}

#### Analysis
\begin{equation*}
X(\omega) = \mathscr{F}\{(x(t)\}= \int_{-\infty}^{\infty} x(t) \hspace{0.1cm} e^{j\omega t} dt
\end{equation*}

The above equation can be written as a function of frequency $f$ instead of $\omega$ and since $f=\frac{2\pi}{\omega}$, the $2\pi$ factor above disappears for $x(t) \leftrightarrow X(f)$. The Fourier Transform has many interesting properties that can be explored [here](http://fourier.eng.hmc.edu/e101/lectures/handout3/node2.html). The Fourier Transform of some important signals can be found [here](http://fourier.eng.hmc.edu/e101/lectures/handout3/node3.html).

Note that the quantity $X(f)$ is a complex function that produces a complex number at each of the frequencies. The magnitude of that complex function is directly related to the spectrum you have been plotting in Audacity of various music signals. Keeping in mind that energy is related to the magnitude squared, the spectrum gives you an idea of how much of the signal energy is present at every frequency. More on the magnitude and phase of the Fourier Transform in the [Appendix](#Appendix).

This idea of investigating how much of a signal's energy (component) is present at distinct frequency values has many applications: in music, if we identify the bass and treble frequencies, then we can boost the frequencies that we care about and suppress the ones we don't. Similary in radio transmission, we can listen to different channels because of this idea that the frequency spectrum can be divided into chucks for each station that we can tune to. More over, if we can identify the frequency of vibrations of earthquakes, we can build structures, like high-rises and bridges such that they don't resonate (have high energy vibrations) at earthquake frequencies. We will look at applications of frequency or spectrum analysis in data compression (reducing the size of data files) as well, where you can think of patterns that repeat very little, as having a small frequency and basically don't provide a lot of information, therefore we can safely get rid of them by 'filtering' those frequencies out, or removing them. 

From signals, we make waves, and from waves we can compute the Sectrum which is a form of a Fourier Transform that uses an alogrithm called the FFT (Fast Fourier Transform). See below for an example for a triangle signal and a square signal with a fundamental frequency at 200 Hz. Note the envelope of both after you run the cells below. In the Triangle wave the amplitude drops from the fundamental as a function $\frac{1}{f^2}$ and for the Square wave as a function of $\frac{1}{f}$. For more complex sounds, these envelops are harder to charactrize though as we saw in class, we often can identify the ADSR phases for musical notes on various instruments. 

```python
signal = thinkdsp.TriangleSignal(200)
wave = signal.make_wave(duration = 0.5, framerate = 10000)
spectrum = wave.make_spectrum()
spectrum.plot(high=3000)
```

```python
signal = thinkdsp.SquareSignal(200)
wave = signal.make_wave(duration = 0.5, framerate = 10000)
spectrum = wave.make_spectrum()
spectrum.plot(high=3000)
```

### How to do this more elegantly?
Alternatively we can use the following `synthesize` function, which we've "defined" below. Pay attention to how you define a funciton. You can hide a lot of repeated code in a function and make for more efficient programming. So, instead of copying the amplitude, frequency pairs into appropriate `CosSignal` functions, our function does it for us! We just pass in `peaks_7` that we defined as the first 7 peaks of the spectrum as an argument to `synthesize`.

Note that this function takes two inputs, it uses a built in function in the ThinkDSP library called SumSignal that sums a list of `CosSignal` or `SinSignal` objects, and returns the y_values for the synthesized signals. There is a `for` loop structure that takes advantage of the repetition in the calcuations we did above. There are a lot more details that we can explore with you if you are interested in unpacking this more. 

```python
def synthesize(peaks, ts):
    components = [thinkdsp.CosSignal(freq, amp) for amp, freq in peaks]
    signal = thinkdsp.SumSignal(*components)
    
    ys = signal.evaluate(ts)
    return ys
```

```python
framerate = 11025

ts = np.linspace(0, 1, framerate)
ys = synthesize(peaks_7, ts)
wave4 = thinkdsp.Wave(ys, ts, framerate)
wave4.apodize()
wave4.plot()
thinkplot.config(xlabel='Time (s)')
wave4.make_audio()
```

`apodize()` removes noise from the *feet* or edges of the signal, it's a smoothing window function. 

Obviously neither sound exactly like the violin note and the envelopes don't look exactly the same. For one thing we are not incorporating phase information (how each tone is phase shifted in relationship to the other tones), and we didn't include all the peaks. Note that in the example right above the synthesize function definition, we used a mix of `Sin` and `Cos` signals at roughly the peaks and frequencies that were given by the peaks function. If you were to only use `Cos` with more accurate numbers like in the synthesize function, you will get the plot above.

You can go from the trignometric form to the exponential form using  Euler's equation, which is often referred to as *the most beautiful formula in mathematics!* 
\begin{equation*}
e^{jx} = cos(x)+jsin(x)
\end{equation*}
And therefore for $x = \pi$, we get Euler's identiy:
\begin{equation*}
e^{j\pi}+1 = 0

The nice thing about the basis functions is that they are **orthogonal**, which means for different $n\neq m$ these functions have nothing in common. Orthogonal *basis* functions like these are very efficient (no waste or overlap) in representing other functions, almost like lego blocks that are so unique that you can't make one of the building blocks out of two other blocks.


**Analysis**

The following analysis equations are the ones where the opposite of synthesis occurs. A Fourier series in this form is like a prism, where a periodic function $x(t)$ with a period $T$ is passed through this Fourier prism to figure out what it is made out of. This is spectral analysis at its simplest form. 

\begin{equation*}
a_0= {2 \over T} \int_{T} x(t)_T \hspace{0.1cm} dt
\end{equation*}
\begin{equation*}
a_n = {2 \over T} \int_{T} x(t)_T \hspace{0.1cm} \cos(n\omega_o t)\hspace{0.1cm} dt  \hspace{1cm} n \neq 0
\end{equation*}
\begin{equation*}
b_n = {2 \over T} \int_{T} x(t)_T \hspace{0.1cm} \sin(n\omega_o t)\hspace{0.1cm} dt  
\end{equation*}

And if you use the complex exponential series form 
\begin{equation*}
c_n = {1 \over T} \int_{T} x(t)_T \hspace{0.1cm} e^{jn\omega_o t}\hspace{0.1cm} dt  
\end{equation*}

Note that if we multiply two functions or signals together and take the integral of the product, the result is a measure of how similar they are, a correlation, as can be seen below ($t$ is a real number). 

\begin{equation*}
\int_{t \in \Re}^{} f(t) g(t) dt
\end{equation*}

| ![similarity](similarity.jpg) | 
|:--:| 
|*From Muller section 2.1.1.2*|

If areas above and below the y-axis are summed (an integration), the result is much larger in the situation in part (a) than in part (b) where most of the areas cancel each other (negative and positive). If the areas completely cancel each other, we call the functions **orthogonal**. So for the orthogonal basis functions we mentioned earlier, the product of any of those functions with different $n$'s you get zero, or the areas under the curve at various times cancel each other, for example:

\begin{equation*}
\int_{-\infty}^{\infty} \cos(n\omega_o t)\hspace{0.1cm} \sin(m\omega_o t) \hspace{0.1cm}dt = 0 \hspace{0.5cm} for \hspace{0.1cm} n \neq m
\end{equation*}

Therefore, $a_n$, $b_n$ and $c_n$ are effectively measuring how much  $\sin(n\omega_o t)$, $\cos(n\omega_o t)$, and$e^{jn\omega_o t}$ is in $x(t)$, or how similar is $x(t)$ to these basis functions. That's why we call those equations the analysis equations, as we are analyzing the function to see how much of each these basis functions for each n is in it.

#### Why the exponential form?

We draw attention to the complex exponential form because the phasor is an important tool in describing various systems in signal processing. ### More about the Fourier Transform
#### Magnitude and Phase

The magnitude of a complex number $c = a +jb = Ae^{j\theta}$, where $a$ is the real part and $b$ is the imaginary part is 

\begin{equation*}
A = (c)(c)^* = (a+jb)(a-jb) = |a+jb| = \sqrt{a^2 + b^2}
\end{equation*}

where $c^*$ is the complex conjugate of $n$ (same real value but opposite sign imaginary)

The phase $\theta$ can be calculated as 

\begin{equation*}
\theta = tan^{-1} \frac{Im\{c\}}{Re\{c\}}
\end{equation*}

Similary for the Fourier Transfer function $X(j\omega)$, a complex function of $j\omega$, the magnitude which determines the relative presence of a sinusoid $e^{j\omega t}$ in $x(t)$, and the the phase which determines how the sinusoids line up relative to one another to form x(t) can be calculated as follows:

\begin{equation*}
|X(j\omega)| = \sqrt{(X(j\omega)(X(j\omega)^*}  \\
\angle X(j\omega) =  tan^{-1} \frac{Im\{X(j\omega)\}}{Re\{X(j\omega)\}}
\end{equation*}

Again note that the magnitude of the Fourier Transform is related to the spectrum you have been plotting in Audacity of various music signals. Keeping in mind that energy is related to the magnitude squared, the spectrum gives you an idea of how much of the signal energy is present at every frequency. 

#### What is negative frequency anyway?
An Audio signal $x(t)$ is a real function of time where there are always two equal-amplitude complex exponentials, rotating in opposite directions, one clock-wise and one counter-clock wise (see spirals above). This results in the real parts combining and the imaginary parts canceling out, leaving only a real sinusoid as the result. For example, for a $\sin$ wave, the spectrum contains 2 spikes, one positive frequency and one negative. Depending on the phase of the two spirals, they could cancel out leaving a purely real sine wave or a real cosine wave (phase shift of $90 ^o$). Note that the negative and positive frequency components are both necessary to produce the real signal. One can show that for a real signal $|X(j\omega)| = |X(-j\omega)|$, resulting in an even function for the magnitude, and an odd function of phase. That's why we only show the spectrum for positive values of frequency, because there is no new information given by the negative fequencies.
